```python
#P value 1.7140303479358558e-05 --> scientific notation
# p is low, null must go: we reject the H0 hypothesis
# conclude that the means attack and defense are not equal
```

```python
#What is the significance level


#The significance level, also denoted as alpha or Î±, is the probability of rejecting the null hypothesis when it is true. 
#For example, a significance level of 0.05 indicates a 5% risk of concluding that a difference exists when there is no actual difference.
#A common value used for alpha is 5% or 0.05. 
#A smaller alpha value suggests a more robust interpretation of the null hypothesis, such as 1% or 0.1%
```

```python
#What are the degrees of freedom of model


#Degrees of freedom of an estimate is the number of independent pieces of information that went into calculating the estimate.
#The most commonly encountered equation to determine degrees of freedom in statistics is df = N-1
```

```python
#error terms


#Each of these two types of error has a specific name.

#Type I Error: The incorrect rejection of a true null hypothesis or a false positive.
#Type II Error: The incorrect failure of rejection of a false null hypothesis or a false negative.
```

```python
#ANOVA test
#to compare the means of multiple groups at the same time
```

```python
data_etching_rate['Etching Rate'] = data_etching_rate.groupby('Etching Rate').cumcount() 

data_etching_rate_pivot = data_etching_rate.pivot(index='Etching Rate', columns='Etching Rate', values='Power')
data_etching_rate_pivot.columns = ['Etching Rate'+str(x) for x in data_etching_rate_pivot.columns.values]
data_etching_rate_pivot.head()
```
